```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error
import matplotlib.pyplot as plt
import seaborn as sns

from statsmodels.tsa.arima.model import ARIMA
from pmdarima import auto_arima

from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from xgboost import XGBRegressor

import keras.models
from keras.models import Sequential
from keras.layers import LSTM, Dense
from sklearn.preprocessing import MinMaxScaler

import warnings
warnings.filterwarnings("ignore")
```

```python
#df = df.reset_index()  # puts 'Date' back as a column
#df.set_index('Date', inplace=True)
#df['Date'] = pd.to_datetime(df['Date'])
df.index = pd.to_datetime(df.index)  # ✅ This works because 'Date' is the index, not a column

# Resample to monthly or weekly if daily data
df = df.resample('M').mean(numeric_only=True)  # or 'W' for weekly

# Fill missing values
df['AQI'] = df['AQI'].interpolate()
```

```python
# Train-test split (80% train, 20% test)
train_size = int(len(df) * 0.8)
train, test = df['AQI'][:train_size], df['AQI'][train_size:]
```

```python
from pmdarima import auto_arima

# Auto ARIMA to find best parameters
model_arima = auto_arima(train, seasonal=False, trace=True)

# Fit model
model = ARIMA(train, order=model_arima.order)
model_fit = model.fit()

# Forecast
forecast_arima = model_fit.forecast(steps=len(test))
```

```python
def create_lag_features(series, lags=12):
    df_lag = pd.DataFrame()
    for i in range(1, lags+1):
        df_lag[f'lag_{i}'] = series.shift(i)
    df_lag['target'] = series.values
    return df_lag.dropna()

lags = 12
lagged_df = create_lag_features(df['AQI'], lags)
X = lagged_df.drop(columns='target')
y = lagged_df['target']

X_train, X_test = X[:train_size-lags], X[train_size-lags:]
y_train, y_test = y[:train_size-lags], y[train_size-lags:]

# Random Forest
rf = RandomForestRegressor()
rf.fit(X_train, y_train)
rf_pred = rf.predict(X_test)

# XGBoost
xgb = XGBRegressor()
xgb.fit(X_train, y_train)
xgb_pred = xgb.predict(X_test)
```

```python
# Scale data
scaler = MinMaxScaler()
scaled = scaler.fit_transform(df[['AQI']])

def prepare_lstm_data(series, timesteps=12):
    X, y = [], []
    for i in range(len(series) - timesteps):
        X.append(series[i:i+timesteps])
        y.append(series[i+timesteps])
    return np.array(X), np.array(y)

X_lstm, y_lstm = prepare_lstm_data(scaled, 12)
X_lstm_train, X_lstm_test = X_lstm[:train_size-12], X_lstm[train_size-12:]
y_lstm_train, y_lstm_test = y_lstm[:train_size-12], y_lstm[train_size-12:]

# Build model
model = Sequential([
    LSTM(50, activation='relu', input_shape=(X_lstm_train.shape[1], 1)),
    Dense(1)
])
model.compile(optimizer='adam', loss='mse')
model.fit(X_lstm_train, y_lstm_train, epochs=20, verbose=0)

# Predict
y_lstm_pred = model.predict(X_lstm_test)
y_lstm_pred = scaler.inverse_transform(y_lstm_pred)
y_lstm_test_true = scaler.inverse_transform(y_lstm_test.reshape(-1, 1))
```

```python
def evaluate(true, pred, model_name):
    print(f"{model_name}")
    print("MAE:", mean_absolute_error(true, pred))
    print("RMSE:", mean_squared_error(true, pred, squared=False))
    print()

evaluate(test, forecast_arima, "ARIMA")
evaluate(y_test, rf_pred, "Random Forest")
evaluate(y_test, xgb_pred, "XGBoost")
evaluate(y_lstm_test_true, y_lstm_pred, "LSTM")
```

```python
# Add coordinates manually
coords = {
    'BTM Layout': (12.9166, 77.6101),
    'Jayanagar': (12.9250, 77.5938),
    'Hebbal': (13.0350, 77.5913),
    'Peenya': (13.0228, 77.5261),
    'City Railway Station': (12.9764, 77.5696)
}

combined_df1['Lat'] = combined_df1['Station'].apply(lambda x: coords[x][0])
combined_df1['Lon'] = combined_df1['Station'].apply(lambda x: coords[x][1])
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
for station in combined_df1['Station'].unique():
    station_df = combined_df1[combined_df1['Station'] == station]
    plt.plot(station_df['Date'], station_df['AQI'], label=station)

plt.title("AQI Trends by Station (2018–2024)")
plt.xlabel("Date")
plt.ylabel("AQI")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
combined_df1['Date'] = pd.to_datetime(combined_df1['Date'], errors='coerce')

# Create a Year-Month column
combined_df1['YearMonth'] = combined_df1['Date'].dt.to_period('M').astype(str)

# Pivot for heatmap
pivot_df = combined_df1.groupby(['Station', 'YearMonth'])['AQI'].mean().unstack()

import seaborn as sns

plt.figure(figsize=(14, 6))
sns.heatmap(pivot_df, cmap='Reds', linewidths=0.5, linecolor='gray', cbar_kws={'label': 'Avg AQI'})
plt.title("Monthly Avg AQI by Station")
plt.xlabel("Month-Year")
plt.ylabel("Station")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Feature Engineering for AQI

```python
combined_df1['Date'] = pd.to_datetime(combined_df1['Date'])
combined_df1['Month'] = combined_df1['Date'].dt.month
combined_df1['Season'] = combined_df1['Month'].map({
    1: 'Winter', 2: 'Winter', 3: 'Summer',
    4: 'Summer', 5: 'Summer', 6: 'Monsoon',
    7: 'Monsoon', 8: 'Monsoon', 9: 'Post-Monsoon',
    10: 'Post-Monsoon', 11: 'Winter', 12: 'Winter'
})

combined_df1.groupby('Season')['AQI'].mean().plot(kind='bar')
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
result = seasonal_decompose(combined_df1['AQI'], model='additive', period=365)
result.plot()
```

```python
## Lagged Effects of Meteorological Variables

for lag in range(1, 8):
    combined_df1[f'temp_lag_{lag}'] = combined_df1['Temperature'].shift(lag)
    combined_df1[f'humidity_lag_{lag}'] = combined_df1['Humidity'].shift(lag)
    combined_df1[f'wind_lag_{lag}'] = combined_df1['WindSpeed'].shift(lag)
```

```python
## Lagged AQI values

for lag in range(1, 8):
    combined_df1[f'AQI_lag_{lag}'] = combined_df1['AQI'].shift(lag)

#combined_df1.corr()['AQI'][df.columns.str.contains('lag')]
```

```python
## Sudden Change Detection (Events, Lockdowns)

combined_df1['Lockdown'] = ((combined_df1['Date'] >= '2020-03-25') & (combined_df1['Date'] <= '2020-06-01')).astype(int)

diwali_dates = ['2019-10-27', '2020-11-14', '2021-11-04']
combined_df1['Diwali'] = combined_df1['Date'].isin(pd.to_datetime(diwali_dates)).astype(int)

combined_df1.groupby('Lockdown')['AQI'].mean()
combined_df1.groupby('Diwali')['AQI'].mean()
```

# Clustering

```python
jayanagar = pd.read_csv("Jayanagar_PM25_2016_2024.csv")
peenya = pd.read_csv("Peenya_PM25_2016_2024.csv")
hebbal = pd.read_csv("Hebbal_PM25_2016_2024.csv")
btm_layout = pd.read_csv("BTM Layout_PM25_2016_2024.csv")
crs = pd.read_csv("City Railway Station_PM25_2016_2024.csv")

df = pd.concat([jayanagar, peenya, hebbal, btm_layout, crs], ignore_index=True)
```

```python
df['Date'] = pd.to_datetime(df['Date'])

# Pivot to get a matrix: rows = locations, columns = time (monthly AQI)
pivot_df = df.pivot(index='Station', columns='Date', values='AQI')

# Fill missing values (e.g., forward fill or mean fill)
pivot_df = pivot_df.fillna(method='ffill', axis=1).fillna(method='bfill', axis=1)
```

```python
## Normalization

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
data_scaled = scaler.fit_transform(pivot_df)

# Keep the location labels
locations = pivot_df.index
locations
```

```python
## K-mean Clustering

from sklearn.cluster import KMeans

# Decide number of clusters (try 3 to start with)
k = 3
kmeans = KMeans(n_clusters=k, random_state=42)
clusters = kmeans.fit_predict(data_scaled)

# Assign cluster labels back to locations
cluster_df = pd.DataFrame({
    'Location': locations,
    'Cluster': clusters
})

print(cluster_df.sort_values(by='Cluster'))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Add cluster labels back to original pivot_df for plotting
pivot_df['Cluster'] = clusters

# Plot average AQI trend for each cluster
plt.figure(figsize=(12, 6))
for cluster_id in range(k):
    cluster_group = pivot_df[pivot_df['Cluster'] == cluster_id].drop('Cluster', axis=1)
    mean_trend = cluster_group.mean()
    plt.plot(mean_trend.index, mean_trend.values, label=f'Cluster {cluster_id}')

plt.title("Average AQI Trend per Cluster")
plt.xlabel("Month")
plt.ylabel("Standardized AQI")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# PCA for 2D visualization

from sklearn.decomposition import PCA

pca = PCA(n_components=2)
components = pca.fit_transform(data_scaled)

plt.figure(figsize=(8, 6))
sns.scatterplot(x=components[:, 0], y=components[:, 1], hue=clusters, palette='Set2')
for i, location in enumerate(locations):
    plt.text(components[i, 0], components[i, 1], location, fontsize=9)

plt.title("Location Clustering by AQI Trends (PCA)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
from scipy.spatial.distance import pdist, squareform
import matplotlib.pyplot as plt
import seaborn as sns

# --- 1. wide table: rows = location, columns = monthly AQI values
X = pivot_df.values      # pivot_df from earlier (no 'Cluster' col this time)

# --- 2. correlation distance matrix
corr_dist = pdist(X, metric=lambda u, v: 1 - pd.Series(u).corr(pd.Series(v)))

# --- 3. hierarchical clustering (Ward or average linkage)
Z = linkage(corr_dist, method='average')

# --- 4. plot dendrogram
plt.figure(figsize=(6, 4))
dendrogram(Z, labels=pivot_df.index, leaf_rotation=90)
plt.title("Location similarity (correlation distance)")
plt.tight_layout()
plt.show()

# --- 5. cut tree into clusters if you still want labels
clusters = fcluster(Z, t=0.4, criterion='distance')   # adjust threshold
cluster_df = pd.DataFrame({'Location': pivot_df.index, 'Cluster': clusters})
```
